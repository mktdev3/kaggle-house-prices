# Exploratory Data Analysis (EDA)
This notebook performs initial data analysis to understand the dataset structure, distribution, and correlations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set_style('whitegrid')
```

```python
# Load data
train_df = pd.read_csv('../data/train.csv')
test_df = pd.read_csv('../data/test.csv')

print(f"Train Shape: {train_df.shape}")
print(f"Test Shape: {test_df.shape}")
```

## Target Variable Analysis: SalePrice


```python
plt.figure(figsize=(10, 6))
sns.histplot(train_df['SalePrice'], kde=True)
plt.title('Distribution of SalePrice')
plt.show()

print(f"Skewness: {train_df['SalePrice'].skew()}")
print(f"Kurtosis: {train_df['SalePrice'].kurt()}")
```

## Missing Values


```python
missing = train_df.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)

plt.figure(figsize=(12, 6))
missing.plot.bar()
plt.title('Missing Values Count')
plt.show()
```

## Correlation Matrix


```python
# Numerical features correlation
numeric_features = train_df.select_dtypes(include=[np.number])
corr = numeric_features.corr()

# Top 10 correlated features with SalePrice
cols = corr.nlargest(11, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(train_df[cols].values.T)
sns.set(font_scale=1.25)
plt.figure(figsize=(12, 12))
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.title('Top 10 Correlated Features with SalePrice')
plt.show()
```
